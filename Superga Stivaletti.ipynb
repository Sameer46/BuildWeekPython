```python
### Stivaletti - Superga
import bs4 
import pandas as pd 
import requests
url = "https://www.superga.com/it/collections/stivaletti-donna?filter.p.m.custom.seasons=Primavera%2FEstate"
page = requests.get( url ) 
soup = bs4.BeautifulSoup( page.text ) 
NameStiv = soup.find_all("div", class_="grid-product__title grid-product__title--heading")
stivname = []
for stiv in NameStiv:
    stivname.append(stiv.text.strip())
```

```python
colstiv = soup.find_all("span",  class_="cb-italic cb-block cb-font-500")
stivcol = []
for stiv in colstiv:
    stivcol.append(stiv.text.strip())
```

```python
stivprice = soup.find_all("div", class_="grid-product__price")
pricestiv= []
pint = []
for price in stivprice:
    price = price.text.strip()
    price = price.replace("\n", "")
    price = price.replace("da","")
    price = price.strip()
    price = price.replace("â‚¬","")
    price = price.replace(",",".")
    pricestiv.append(float(price))
    pint.append(float(price))
```

```python
data = {"Nome" : stivname, "IntPrice" : pint, "ScontPrice" : pricestiv ,"Colore" : stivcol, "Categoria" : "Stivaletti", "Gender": "F"}
```

```python
SupStivF = pd.DataFrame ( data )
```
