```python
### Scarpe Basse - Superga
import bs4
import pandas as pd
import requests
url = "https://www.superga.com/it/collections/scarpe-basse-donna?filter.p.m.custom.seasons=Estate&filter.p.m.custom.seasons=Primavera%2FEstate"
page = requests.get( url )
soup = bs4.BeautifulSoup( page.text )
Nameslow = soup.find_all ("div", class_="grid-product__title grid-product__title--heading")
slow = []
for slow in Nameslow :
    slow.append( slow.text.strip())
```

```python
colslow = soup.find_all("div", class_="btn btn--body btn--circle js-photoswipe__zoom product__photo-zoom")
slowcol = []
for slow in colslow:
    slowcol.append(slow.text.strip())
    
```

```python
slowprice = soup.find_all("div", class_="grid-product__price")
priceslow= []
pint = []
for price in slowprice:
    price = price.text.strip()
    price = price.replace("\n", "")
    price = price.replace("da","")
    price = price.strip()
    price = price.replace("â‚¬","")
    price = price.replace(",",".")
    priceslow.append(float(price))
    pint.append(float(price))
```

```python
data = {"Nome" : Nameslow, "IntPrice" : pint, "ScontPrice" : priceslow ,"Colore" : slowcol, "Categoria" : "Scarpe Basse", "Gender": "F"}
```

```python
SupSlowF = pd.DataFrame ( data )
```
